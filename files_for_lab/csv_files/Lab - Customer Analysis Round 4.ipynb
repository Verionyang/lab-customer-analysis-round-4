```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
data = pd.read_csv('marketing_customer_analysis.csv')

cols = []
for column in data.columns:
    cols.append(column.lower().capitalize().replace(" ","_"))

data.columns=cols
    
display(data.head())
display(data.shape)
```

```python
data.dtypes
```

```python
data_numeric = data.select_dtypes(include = np.number)
data_cata = data.select_dtypes(include = np.object)

display(data_numeric.head())
display(data_cata.head())
```

```python
fig, ax = plt.subplots(2,4,figsize = (10,5))
sns.distplot(data_numeric['Customer_lifetime_value'],ax=ax[0][0])
sns.distplot(data_numeric['Income'],ax=ax[0][1])
sns.distplot(data_numeric['Monthly_premium_auto'],ax=ax[0][2])
sns.distplot(data_numeric['Months_since_last_claim'],ax=ax[0][3])
sns.distplot(data_numeric['Months_since_policy_inception'],ax=ax[1][0])
sns.distplot(data_numeric['Number_of_open_complaints'],ax=ax[1][1])
sns.distplot(data_numeric['Number_of_policies'],ax=ax[1][2])
sns.distplot(data_numeric['Total_claim_amount'],ax=ax[1][3])
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,4,figsize = (10,5))
sns.distplot(data_numeric['Customer_lifetime_value'],ax=ax[0][0])
sns.distplot(data_numeric['Income'],ax=ax[0][1])
sns.distplot(data_numeric['Monthly_premium_auto'],ax=ax[0][2])
sns.distplot(data_numeric['Months_since_last_claim'],ax=ax[0][3])
sns.distplot(data_numeric['Months_since_policy_inception'],ax=ax[1][0])
sns.distplot(data_numeric['Number_of_open_complaints'],ax=ax[1][1])
sns.distplot(data_numeric['Number_of_policies'],ax=ax[1][2])
sns.distplot(data_numeric['Total_claim_amount'],ax=ax[1][3])
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,4,figsize = (10,5))
ax[0,0].hist(data_numeric['Customer_lifetime_value'], bins=10)
ax[0,1].hist(data_numeric['Income'],bins=10)
ax[0,2].hist(data_numeric['Monthly_premium_auto'],bins=10)
ax[0,3].hist(data_numeric['Months_since_last_claim'],bins=10)
ax[1,0].hist(data_numeric['Months_since_policy_inception'],bins=10)
ax[1,1].hist(data_numeric['Number_of_open_complaints'],bins=10)
ax[1,2].hist(data_numeric['Number_of_policies'],bins=10)
ax[1,3].hist(data_numeric['Total_claim_amount'],bins=10)
plt.tight_layout()
plt.show()
```

Only Income and Months since policy inception follow the shape of a normal distribution. The rest of the numeric values doesn't.
